```
# No 1

import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow

print(' Gamma Correction pada citra ') 
print('----------------------------------') 
try: 
  gamma = int(input('Masukkan nilai Gamma: ')) 
except ValueError: 
  print('Error, not a number')

image = cv.imread('/content/drive/MyDrive/PVCK/female.tiff')
gamma_image = np.zeros(image.shape, image.dtype)

# Akses per pixel
for y in range(image.shape[0]):
  for x in range(image.shape[1]):
    for c in range(image.shape[2]):
      gamma_image[y,x,c] = np.clip(225 * pow(image[y,x,c]/255, (1/gamma)), 0, 255)

final_frame = cv.hconcat((image, gamma_image))
cv2_imshow(final_frame)
```

```
# No 4

import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

image = cv.imread('/content/drive/MyDrive/PVCK/teeth.jpg')
mask = np.zeros(image.shape, dtype=np.uint8)
mask = cv.rectangle(mask, pt1=(320,595), pt2=(800,1025), color=(255,255,255), thickness = -1)

result = image & mask

mask1 = cv.rectangle(mask,pt1=(405,5), pt2=(850,455), color=(255,255,255), thickness=-1)

result = image & mask1

final_frame = cv.hconcat((image, result))
cv2_imshow(final_frame)
```

```
# 5 - OR
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

image = cv.imread('/content/drive/MyDrive/PVCK/teeth.jpg')
mask_or = np.zeros(image.shape, dtype=np.uint8)

mask_or = cv.rectangle(mask_or, pt1=(320,595), pt2=(800,1025), color=(255,255,255), thickness=-1)
result = image | mask_or

mask_or1 = cv.rectangle(mask_or, pt1=(405,5), pt2=(850,455), color=(255,255,255), thickness=-1)
result_or1 = image | mask_or1

final_frame = cv.hconcat((image, result_or1))
cv2_imshow(final_frame)
```
